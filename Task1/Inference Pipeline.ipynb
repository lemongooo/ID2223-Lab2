```python
import gradio as gr
import torch
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="lemongooooo/lora_data_0",
    max_seq_length=2048,
    dtype=torch.float16,
    load_in_4bit=True
)
model = FastLanguageModel.for_inference(model)


def generate_recipe(ingredients):
    prompt = f"You are a professional chef. Based on the following ingredients: {ingredients}, create a delicious recipe including preparation steps and tips."

    inputs = tokenizer.apply_chat_template(
        [{"role": "user", "content": prompt}],
        add_generation_prompt=True,
        return_tensors="pt"
    ).to(model.device)

    outputs = model.generate(
        inputs,
        max_new_tokens=512,
        temperature=0.7,
        do_sample=True,
    )

    response = tokenizer.decode(outputs[0][len(inputs[0]):], skip_special_tokens=True)
    return response


demo = gr.Interface(
    fn=generate_recipe,
    inputs=gr.Textbox(label="Enter ingredients (comma-separated)", placeholder="e.g., chicken, onion, garlic"),
    outputs=gr.Textbox(label="Generated Recipe"),
    title="AI Recipe Generator",
    description="Enter ingredients, and the chatbot will generate a recipe for you!",
    theme=gr.themes.Soft()
)


demo.launch(share=True)
```
